```python
#Function showing the mean survival for a numerical category
def mean_survival(feature):
    try:
        '''For loop through categorical variables'''
        for cat in categorical:
            '''Checking feature isn't categorical if so raises value error'''
            if feature in cat:
                raise ValueError
        mean = data.groupby('overall_survival')[feature].mean()
        print(f'The mean survival for {feature} is:')
        print(mean)
    except ValueError:
        print('Enter a numerical feature')

#Using the function to find mean positive lymph nodes of survival
mean_survival('lymph_nodes_examined_positive')
```

```python
#Using the function to find mean age of survival
mean_survival('age_at_diagnosis')
```

```python
#Testing a categorical feature
mean_survival('cellularity')
```

When inputting positive lymph nodes, the function shows that a higher count of positive lymph nodes on average is more likely to lead to not surviving. With a mean of 1.22 positive lymoh nodes for survivng and 2.57 for those that did not survive. This function can be used for any numerical feature, if a categorical feature is entered it will respond with a value error. This function is useful as can easily see if a feature has an affect on survival, for example when inputting age, can see that the means show that an older age of 64.4 is more likely to not survive that a younger women of age 56.5.

```python
#Importing numpy library
import numpy as np

#Function helping to see if there is a correlation between two numerical features
def correlation(feature1, feature2):
    try:
        cor = np.corrcoef(data[feature1].values, data[feature2].values)[0,1]
        print(f'The correlation between {feature1} and {feature2} is {cor:.3f}')
    except ValueError:
        print('Enter numerical features')
    '''Checking to see if correlation or not'''
    if cor >= 0.5 or cor <= -0.5:
        print('Features show a correlation')
    else:
        print('No significant correlation is shown')
```

```python
#Using the function to see correlation between two features
correlation('age_at_diagnosis', 'lymph_nodes_examined_positive')
```

This is a very low correlation so the two features provide different information. This is a useful function as if two features have a high correlation (where output is close to 1 or -1) then no new information is being given. Instead, it can put more weight on the two features and show to have more significance than other potentially more significant features. Therefore, if two features show to have a high correlation then one of the features should be dropped.

A one sample t-test is a statistical method which can be used to see if the mean of a single sample is significantly different to the population mean. This is useful as it can help determine whether differences have occured by chance or reflect a true difference. For example, according to American Cancer Society (2024) the population mean age for breast cancer in the USA is 62, where this data is collected in the UK we can determine if there is a true difference between the two countries. After testing (shown below), the null hypothesis can be rejected, due to the p-value being less than 0.05, concluding the mean is significantly different in the UK with a confidence interval of the true mean being between 60.51 and 61.68.

```python
#Importing stats function
import scipy.stats as stats

#One sample t-test function
def one_sample_ttest(feature1, population_mean):
    stat = stats.ttest_1samp(data[feature1], population_mean)
    ci = stat.confidence_interval(confidence_level=0.95)
    print(f'The T-stat is {stat.statistic:.2f} with p value {stat.pvalue:.2f}')
    print(f'The confidence interval is between {ci.low:.2f} and {ci.high:.2f}')

#One sample t-test for age
one_sample_ttest('age_at_diagnosis', 62)
```

```python
#Turning categorical variables which have order into numerical
from sklearn.preprocessing import OrdinalEncoder

def ordinal(data, feature, *args):
    model = OrdinalEncoder(categories = [list(args)])
    data[feature]= model.fit_transform(data[[feature]])
```

```python
#Turning tumor size category and cellularity into ordinal numerical values
ordinal(data, 'tumor_size_category', 'Small', 'Medium', 'Large', np.nan)
ordinal(data, 'cellularity', 'Low', 'Moderate', 'High', np.nan)
```

This has sorted the categories with order. The other categorical variables can then be sorted after missing values have been filled in.

The functions used can be useful to investigate the data further. Often large data can have an array of different features which aren't all necessary in order to find good results. In fact, by keeping features which do not provide important information can create noise within the dataset, which can cause models to overfit. 

The first function, shows the mean of survival for different numerical features. This funtion allows us to see if there are differences between if a patient survives or not for the inputted feature. If the feature gives a similar mean for surviving and not surviving then the feature can be removed, as it is not providing useful information to help predict whether a patient is going to survive.

The second function investigates correlation, if two features have a high corrrelation then more investigation should be done to see whether both features are needed. 

The third function looks at how reliable the data is. Finding population means of different features we can then input this into the function and see if they are similar to the mean within the data set. For example, if the population mean for age of diagnosis in the UK was 60 and the confidence interval was between 20-25, this would raise questions on whether this dataset is an accurate representation of the women around the UK with breast cancer. 

The final feature allows for categorical features which have order to be transforned into numerical values keeping there order.

As there is a large distribution I have decided to fill the missing values for tumour size with the median so, the filled values are not skewed from the few high values within the feature. 

For tumor stage there are 500 values missing, tumor stage is a huge factor into survival rates. I do not want to fill this with the mean or median distorting the results. Therefore although it is losing data the data is large enough to still find useful information so have deciding to drop the rows which are missing tumor stage.

```python
#Filling tumor size with median
data['tumor_size'] = data['tumor_size'].fillna(data['tumor_size'].median())
```

```python
#Dropping tumor stage
data = data.dropna()
#Resetting index
data = data.reset_index(drop=True)
```

Breast isn't a type of cancer so have replaced these values with the mode of cancer type.

```python
#Sorting breast category
miscategorised = []

mode = data['cancer_type_detailed'].mode()[0]

#Appending the index of rows conataining category Breast into a list
for row in range(len(data)):
    if data.loc[row, 'cancer_type_detailed']=='Breast':
        miscategorised.append(row)

#Replacing breast with the mode
for value in miscategorised:
    data.loc[value, 'cancer_type_detailed'] = mode
```

When inspecting outliers, it is importnt to consider that some cases are rare, and do not want to lose information which holds value. 

Firstly with age there seems to be one data point which is someone who is aged 21, this is a rare case however not far away from the lower quartile range so do not want to remove. 

The boxplot for positive lymph nodes has detected outliers however, these outliers are more than likely true results. I do not want the data holding extreme values which can disproportion the rest of the data. Therefore, I am replacing the large entries to 25 positive lymph nodes. This keeps those entries as a high amount without being substantially more then the rest of the data.

(2016) tumors which are over 100mm are considered rare tumours for the generilization of the data I have decided to remove anything over this size. However, there were none so removed tumors larger that 100mm.

When the following cell is ran a window will open which will show four different buttons. These buttons once clicked will each show a different violin plot. The buttons are coloured to match the plots for aesthetics.

```python
#Functions to display violin plot
#Violin plot with tumor stage
def tumor_stage():
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(x='tumor_stage', y='age_at_diagnosis', data=data, hue='overall_survival', 
               split=True, palette='husl')
    
    ax.set_title("Distributions of Age and Survival Against Tumor Stage", fontsize=14)
    ax.set_xlabel("Tumor Stage", fontsize=12)
    ax.set_ylabel("Age", fontsize=12)
    display_plot(fig)

#Violin plot with which breast
def which_breast():
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(x='primary_tumor_laterality', y='age_at_diagnosis', data=data, hue='overall_survival', 
               split=True, palette='husl')
    
    ax.set_title("Distributions of Age and Survival Against Primary Laterality", fontsize=14)
    ax.set_xlabel("Primary Tumor Laterality", fontsize=12)
    ax.set_ylabel("Age", fontsize=12)
    display_plot(fig)

#Violin plot with Chemo
def chemotherapy():
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(x='chemotherapy', y='age_at_diagnosis', data=data, hue='overall_survival', 
               split=True, palette='husl')
    
    ax.set_title("Distributions of Age and Survival Against Chemotherapy", fontsize=14)
    ax.set_xlabel("Chemotherapy", fontsize=12)
    ax.set_ylabel("Age", fontsize=12)
    display_plot(fig)

#Violin plot with tumor size
def tumor_size():
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.violinplot(x='tumor_size_category', y='age_at_diagnosis', data=data, hue='overall_survival', 
               split=True, palette='husl')
    
    ax.set_title("Distributions of Age and Survival Against Tumor Size", fontsize=14)
    ax.set_xlabel("Tumor Size", fontsize=12)
    ax.set_ylabel("Age", fontsize=12)
    display_plot(fig)


# Function to display plot in the Tkinter window
def display_plot(fig):
    for widget in frame_plot.winfo_children():
        widget.destroy()  # Clear previous plot
    canvas = FigureCanvasTkAgg(fig, master=frame_plot)
    canvas.draw()
    canvas.get_tk_widget().pack()


# Create the main Tkinter window
root = tk.Tk()
root.title("Violin Plots Displaying Overall Survival with Different Features Against Age")
root.geometry("1000x600")

# Create a frame to hold the buttons on the left
frame_buttons = tk.Frame(root)
frame_buttons.pack(side="left", padx=20, pady=20)

# Create a frame to hold the plot area
frame_plot = tk.Frame(root)
frame_plot.pack(side="right", padx=20, pady=20, fill="both", expand=True)

# Add buttons for different plots in the left frame
stage_button = tk.Button(frame_buttons, text="Tumor Stage", command=tumor_stage, font=("Arial", 20), fg="#FF6666")
stage_button.pack(pady=5, fill="x")

side_button = tk.Button(frame_buttons, text="Breat Tumor Laterality", command=which_breast, font=("Arial", 20), fg='#009999')
side_button.pack(pady=5, fill="x")

chemo_button = tk.Button(frame_buttons, text="Chemotherapy", command=chemotherapy, font=("Arial", 20), fg='#FF6666')
chemo_button.pack(pady=5, fill="x")

size_button = tk.Button(frame_buttons, text="Tumor Size", command=tumor_size, font=("Arial", 20), fg='#009999')
size_button.pack(pady=5, fill="x")

# Run the application
root.mainloop()
```
